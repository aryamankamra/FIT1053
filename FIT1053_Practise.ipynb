```python
'''
Programs for finding a fibonnaci sequience iteratively and recursively. 
'''
def fibo(n):
    results = []
    a, b = 0, 1
    while len(results) < n:
        results.append(a)
        a, b = b, a + b
    return results

def fibo_rec_nth(n):#gives the nth fib value
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return  fibo_rec_nth(n-1)+fibo_rec_nth(n-2)

def fibo_rec(n):# Returns a list of len(n) of all fibo values. 
    res = []
    for i in range(n):
        res.append(fibo_rec_nth(i))
    return res
print(fibo_rec(10))
```

```python
def seive_eatosthenes(start, stop):
    boolArr = [True for i in range(stop+1)]
    n = 2
    while n**2 <= stop:
        if boolArr[n] == True:
            for item in range(n**2, stop+1, n):
                boolArr[item] = False
        n += 1
    prime_arr = [elem for elem in range(start, stop+1) 
                if boolArr[elem] == True]
    return prime_arr
print(seive_eatosthenes(0, 10))
        
```
