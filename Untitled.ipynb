```python
def fibo(n):
    results = []
    a, b = 0, 1
    while len(results) < n:
        results.append(a)
        a, b = b, a + b
    return results

def fibo_rec_nth(n):#gives the nth fib value
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return  fibo_rec_nth(n-1)+fibo_rec_nth(n-2)

def fibo_rec(n):# Returns a list of len(n) of all fibo values. 
    res = []
    for i in range(n):
        res.append(fibo_rec_nth(i))
    return res
print(fibo_rec(10))
```
